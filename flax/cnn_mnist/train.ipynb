```python
import jax
import jax.numpy as jnp
from flax import nnx

from dataset import load_mnist, DataLoader
from cnn_mnist import Classifier, train  # where CNN is defined

# Initialize model using dummy input
rng = rngs = nnx.Rngs(0)
model = Classifier(rngs=rng)
```

```python
nnx.display(model)
```

```python
data, targets = load_mnist(split="train")

# Create the dataloader
dataloader = DataLoader(
    data=data, targets=targets, return_labels=True, batch_size=64, shuffle=True
)
```

```python
loss_curve = train(model, dataloader, epochs=10, lr=1e-3)
```

```python
import matplotlib.pyplot as plt

plt.plot(loss_curve, marker="o")
plt.title("Training Loss over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```
